```python
# Basic Concatenation 
pd.concat([inv_jan,inv_feb,inv_mar])
```

```python
# Ignore the index
pd.concat([inv_jan,inv_feb,inv_mar], ignore_index=True)
```

```python
# Setting labels to original tables
pd.concat([inv_jan,inv_feb,inv_mar], ignore_index=True, keys=['jan','feb','mar'])
```

```python
# concatenate tables with different column names 

# 무엇으로 sort되다면, 젤 처음 인덱스값으로
pd.concat([inv_jan, inv_feb], sort=True)
```

```python
# 동시에 있는 컬럼만 합병됨
pd.concat([inv_jan, inv_feb], join='inner')
```

## # using append method
- `concat()`method 의 간편화된 버전
- `ignore_index`, `sort` 를 지원
- `keys`, `join`은 지원하지 않음
- 항상 조인은 outer로 됨

```python
inv_jan.append([inv_feb,inv_mar],
               ignore_index=True, 
               sort=True)
```
